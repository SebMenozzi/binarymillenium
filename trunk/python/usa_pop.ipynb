```python
def proc_cluster(clusters, cluster_dat, im):
    
    clusters[:,2] = 0
    num = clusters.shape[0]
    size = cluster_dat.shape
    for x in range(size[0]):
        for y in range(size[1]):
            min_range = 1e8
            min_ind = 0
            if (im[x,y] == 0):
                continue
                
            for i in range(num):
                dx = x - clusters[i,0]
                dy = y - clusters[i,1]
                dist = dx*dx + dy*dy
                weight = clusters[i,3]
                if (weight <= 0.1):
                    weight = 0.1
                if (weight >= 10.0):
                    weight = 10.0
                
                dist += weight*100
                if dist < min_range:
                    min_range = dist
                    min_ind = i
                    
            cluster_dat[x,y] = min_ind
            clusters[min_ind,2] += im[x,y]
    
    # update weights
    avg_pop= total_pop/num
    for i in range(1,num):
        pop = clusters[i,2] 
        weight = pop/avg_pop
        clusters[i,3] =weight
        #print i, weight
    #update centers

    return (clusters, cluster_dat) 
```

```python
num = 49

avg_pop = total_pop/num

# x,y, population, weight
orig_clusters = zeros((num, 4))
#orig_clusters[:,3] = 1.0

orig_cluster_dat = zeros(size)

#initialize clusters

for i in range(num):
    orig_clusters[i, 0] = rand()*size[0]
    orig_clusters[i, 1] = rand()*size[1]
```

```python
clusters = copy(orig_clusters)
cluster_dat = copy(orig_cluster_dat)
im2 = copy(im)

(clusters, cluster_dat) = proc_cluster(clusters, cluster_dat, im2)
figure(1, figsize=(15,15))
imshow(transpose(cluster_dat), interpolation='nearest')
figure(2, figsize=(5,5))
#imshow(transpose(clusters[cluster_dat, 2]), interpolation='nearest')
plot(clusters[:,2])
```
